```python
import numpy as np

min = 10
max = 17
weights = np.array([0.15, -0.10, -0.08, 0.02, 0.13, 0.07])
readings = np.array([[67, 0.7, 0.01, 29.6, 4, 1],[85, 0.3, 0.00, 30.2, 20, 2]])


for reading in range(len(readings)):
    tmp = readings[reading].dot(weights)
    print(tmp)

    if(tmp > 12):
        print("Increased Fire Danger")
    else:
        print("Stagnant Fire Danger")
    
    outputMinusMin = np.array([min, tmp]).dot(np.array([-1, 1]))
    output = np.array(outputMinusMin).dot(np.array([1/(max - min)]))
    print(f"{output[0] * 100:.2f}%")
```
