# What's the Simplest Form of Neural Learning?

### Learning using the Hot and Cold Method

```python
#1 An Empty Network
weight = 0.1
lr = 0.01

def neural_network(input, weight):
    pred = input * weight
    return pred

#2 Predict
input = 8.5     #num_toes
WorL = 1        #True output

pred = neural_network(input, weight)
mse = (pred - WorL)**2
print(mse)

#3 Compare Up
pred_up = neural_network(input, weight + lr)
mse_up = (pred_up - WorL)**2
print(mse_up)

#4 Compare Down
pred_down = neural_network(input, weight - lr)
mse_down = (pred_down - WorL)**2
print(mse_down)
```

# Hot and Cold Learning

```python
weight = 0.5
input = 0.5
goal_pred = 0.8

lr = 0.01



for i in range(1000):
    pred = neural_network(input, weight)
    mse = (pred - goal_pred)**2
    print(f'Error: {mse},   Prediction: {pred}')


    #Hot & Cold Method
    #Compare Up
    pred_up = neural_network(input, weight + lr)
    mse_up = (pred_up - goal_pred)**2

    #Compare Down
    pred_down = neural_network(input, weight - lr)
    mse_down = (pred_down - goal_pred)**2

    if(mse_up < mse_down):
        weight += lr
    else:
        weight -= lr

def neural_network(input, weight):
    pred = input * weight
    return pred
```

# Calculating Both Direction and Amount from Error

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for i in range(20):
    pred = input * weight
    mse = (pred - goal_pred)**2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount

    print(f"Mean Squared Error(MSE): {mse}, Prediction: {pred}")
```

# One Iteration of Gradient Descent

```python
weight = 0.1
alpha = 0.01
def neural_network(input, weight):
    pred = input * weight
    return pred

num_toes = 8.5  #Input
WorL = 1        #True Outcome

pred = neural_network(num_toes, weight)
mse = (pred - WorL)**2

#2 Calculate Node Delta
delta = pred - WorL

#3 Learn by updating the weight
weight_delta = delta * num_toes

weight -= weight_delta * alpha

print(weight)
```

# Learning is Just Reducing Error



```python
weight = 0.0
goal_pred = 0.8
input = 0.5

for i in range(4):
    pred = input * weight
    mse = (pred - goal_pred)**2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight = weight - weight_delta  #Notice no alpha
    print(f"MSE: {mse}, prediciton: {pred}")
```

# Let's Watch Several Steps of Learning
